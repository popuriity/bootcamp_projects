## Get all data

```python
url = "https://catfact.ninja/breeds"

breeds = []
countries = []
origins = []
coats = []
patterns = []

for i in range(1,5):
    # send request to each page
    response = requests.get(f"{url}?page={i}")
    result = response.json()
    
    # get only data key
    data = result['data']
    
    for j in range(0, len(data)):
        # convert to series
        breeds.append(data[j]['breed'])
        countries.append(data[j]['country'])
        origins.append(data[j]['origin'])
        coats.append(data[j]['coat'])
        patterns.append(data[j]['pattern'])
    
    # stop 2 seconds before go to next page
    time.sleep(2)

cat_df = pd.DataFrame({
    'breed': breeds,
    'country': countries,
    'origin': origins,
    'coat': coats,
    'pattern': patterns
})

cat_df
```

Even if `cat_df.info()` shows that all columns has no missing values, there are white spaces in some data point. All data type are object, so I replace these white spaces with `NA`.

```python
cat_df = cat_df.replace("", np.nan)
cat_df.tail(5)
```

```python
# The number of breeds on each of pattern
cat_df["pattern"].value_counts()
```

```python
cat_origin_coat = cat_df.groupby('origin')['coat'].value_counts().reset_index(name = 'count')
cat_origin_coat
```

```python
# find cat breeds that origin is from Crossbred and its coat is short.
cat_df.query(' origin == "Crossbred" & coat == "Short" ')
```
